```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt

china = gpd.read_file("gadm41_CHN_shp/gadm41_CHN_1.shp")

provinces = [
    "Hebei", "Shanxi", "Liaoning", "Jilin", "Heilongjiang",
    "Jiangsu", "Zhejiang", "Anhui", "Fujian", "Jiangxi",
    "Shandong", "Henan", "Hubei", "Hunan", "Guangdong",
    "Hainan", "Sichuan", "Guizhou", "Gansu", "Qinghai"
]

aqi_values = [
    42.4, 45, 29, 17.8, 32.8, 24.5, 22.3, 41, 18, 14,
    34.9, 36.7, 49.9, 14, 22.6, 4, 38.8, 28, 20, 29
]


cancer_rates = [
    37.1, 31.3, 61.5, 52.4, 70.7, 44.5, 51.1, 38.4, 37.6, 40.5,
    46.7, 37.9, 52.5, 40.9, 44.5, 34.4, 50, 36, 20.8, 30.1
]
data_df = pd.DataFrame({
    'NAME_1': provinces,
    'AQI': aqi_values,
    'CancerRate': cancer_rates
})

china = china.merge(data_df, on='NAME_1')

fig, axes = plt.subplots(1, 2, figsize=(18, 10))

china.plot(
    column='AQI',
    cmap='coolwarm',
    linewidth=0.8,
    edgecolor='black',
    legend=True,
    ax=axes[0]
)
axes[0].set_title("Air Quality Index by Province", fontsize=20)
axes[0].axis('off')

china.plot(
    column='CancerRate',
    cmap='Reds',
    linewidth=0.8,
    edgecolor='black',
    legend=True,
    ax=axes[1]
)
axes[1].set_title("Cancer Rates by Province",fontsize=20)
axes[1].axis('off')

plt.tight_layout()
plt.show()
```
