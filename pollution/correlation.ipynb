```python
pollution_index =  [11.22, 7.72, 8.52, 9.36, 11.42, 10.68, 10.78, 10, 7.5, 8.56]
lung_cancer = [1.79, 2.184, 2.1, 2.438, 1.983, 2.319, 2.923, 2.673, 2.75, 2.687]
county = ["Santa Clara", "Santa Cruz", "San Mateo", "San Francisco", "Alameda","Contra Costa","Solano" ,"Napa","Sonoma" ,"Marin" 
]

from scipy.stats import pearsonr

r, p_value = pearsonr(pollution_index, lung_cancer)
print(f"Correlation: {r}, P-value: {p_value}")
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.regplot(x=pollution_index, y=lung_cancer, line_kws={"color": "red"})
plt.xlabel("Pollution Index")
plt.ylabel("Lung Cancer Rates")
plt.title("Correlation Between Pollution Index and Lung Cancer")
plt.grid(True)
plt.show()
```
