```python
import pandas as pd
```

```python
parameters = [
    "Aluminum", "Antimony", "Arsenic", "Barium", "Beryllium", "Bromate", "Cadmium",
    "Chromium", "Fluoride", "Mecrucy", "Nickel", "Nitrate", "Nitrite", "Perchlorate",
    "Selenium", "Thallium", "Halocetic Acids", "Trihalomethanes", "Lead", "Copper"
]

limits_list = [
    0.6, 0.001, 0.0, 2, 0.00008, 0.0001, 0.00004, 0.1, 0.7, 0.00063,
    0.012, 10, 1, 0.001, 0.005, 0.0001, 0.06, 0.00032, 0.0, 0.3
]

weights = {
    "Aluminum": 1, "Antimony": 2, "Arsenic": 4, "Barium": 2, "Beryllium": 3,
    "Bromate": 3, "Cadmium": 4, "Chromium": 3, "Fluoride": 2, "Mercury": 5,
    "Nickel": 2, "Nitrate": 2, "Nitrite": 3, "Perchlorate": 3, "Selenium": 3,
    "Thallium": 4, "Haloacetic Acids": 3, "Trihalomethanes": 3, "Lead": 5,
    "Copper": 2
}

standard_limits = dict(zip(parameters, limits_list))
```

```python
wqi = []
dataset = pd.read_excel("/Users/eddie/Downloads/Water Quality Data.xlsx")
print (dataset)
```

```python
def calc_sub_index(row, param):
    measured = row[param]
    standard = standard_limits[param]
    return max(0, (1 - measured / standard) * 100)
```

```python
def calculate_wqi(row):
    subindices = []
    total_weight = 0
    for param in standard_limits:
        if pd.isna(row.get(param)):
            continue
        sub_index = calc_sub_index(row, param)
        weight = weights[param]
        subindices.append(sub_index * weight)
        total_weight += weight
    return sum(subindices) / total_weight if total_weight > 0 else None
```

```python
dataset["WQI"] = dataset.apply(calculate_wqi, axis=1)
print(dataset)
```
