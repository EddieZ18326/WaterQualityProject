```python
water_quality =  [75.688, 68.948876,72.13962857,63.2111,70.8888,71.475253, 71.1763, 73.108667, 73.9704 ,70.582]
poverty_rate = [7.6, 11.2, 7.3 ,11.9, 9.2,8.3,9.7,8.9,8.6,9.1]
county = ["Santa Clara", "Santa Cruz", "San Mateo", "San Francisco", "Alameda"]
from scipy.stats import pearsonr



r, p_value = pearsonr(water_quality, poverty_rate)
print(f"Correlation: {r}, P-value: {p_value}")
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.regplot(x=poverty_rate, y=water_quality, line_kws={"color": "red"})
plt.xlabel("Poverty Rate")
plt.ylabel("Water Quality Index")
plt.title("Correlation Between Poverty Rate and Water Quality")
plt.grid(True)
plt.show()
```
